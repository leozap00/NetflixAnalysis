```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/Netflix.csv')

df['date_added'] = pd.to_datetime(df['date_added'], format="%B %d, %Y", errors='coerce')

df = df.dropna(subset=['date_added'])

df['year_added'] = df['date_added'].dt.year

titles_per_year = df['year_added'].value_counts().sort_index()

titles_per_year.plot(kind = 'bar', figsize = (10,5))
plt.title("Titoli aggiunti su Netflix per anno")
plt.xlabel('Anno di aggiunta')
plt.ylabel('titoli')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data/Netflix.csv')

df['date_added'] = pd.to_datetime(df['date_added'], format="%B %d, %Y", errors='coerce')

df = df.dropna(subset=['date_added'])

df['month_added'] = df['date_added'].dt.month

titles_per_month = df['month_added'].value_counts().sort_index()

titles_per_month.plot(kind = 'bar', figsize = (10,5))
plt.title("Titoli aggiunti su Netflix al mese")
plt.xlabel('Mese')
plt.xticks(range(12), 
           ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
           rotation=45)
plt.ylabel('titoli')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('data/Netflix.csv')

df['date_added'] = pd.to_datetime(df['date_added'], format="%B %d, %Y", errors='coerce')

df = df.dropna(subset=['date_added'])

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

monthly_by_year = df.groupby(['year_added', 'month_added']).size().unstack()

years_compare= [2019, 2020, 2021]
monthly_by_year_filtered = monthly_by_year.loc[monthly_by_year.index.isin(years_compare)]

mesi = np.arange(1, 13)
width = 0.25 

plt.figure(figsize = (14,6))
for i, year in enumerate(monthly_by_year_filtered.index):
    plt.bar(
        mesi + (i - 1) * width,
        monthly_by_year_filtered.loc[year],
        width=width,
        label=str(year)
    )
plt.title("Titoli aggiunti su Netflix negli ultimi 3 anni")
plt.xlabel('Mese')
plt.xticks(mesi, 
           ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
           rotation=45)
plt.ylabel('titoli')

plt.legend(title="Anno")
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```
